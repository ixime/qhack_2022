```python
tmp = pd.read_csv("../datasets/cell-cycle-regulated-genes-microarray/S-EPMC25624/smallest.csv")
ds_smaller = tmp[["gene_id"]].rename(columns={"gene_id": "NAME"})
tmp = pd.read_csv("../datasets/cell-cycle-regulated-genes-microarray/S-EPMC25624/small_25.csv")
ds_small = tmp[["gene_id"]].rename(columns={"gene_id": "NAME"})
tmp = pd.read_csv("../datasets/cell-cycle-regulated-genes-microarray/S-EPMC25624/large_94.csv")
ds_large = tmp[["gene_id"]].rename(columns={"gene_id": "NAME"})
```

```python
#elu experiment
ds1 = pd.DataFrame(data={'T':[i for i in range(9) for j in range(6178)]})
ds1['NAME'] = np.concatenate([ds_full[['NAME']].values for f in range(9)])
ds1['GENE'] = np.concatenate([ds_full[['GENE']].values for f in range(9)])
ds1['t0'] = ds_full[['elu_120min','elu_150min','elu_180min','elu_210min','elu_240min','elu_270min','elu_300min','elu_330min','elu_360min']].melt()[['value']]
ds1['t1'] = ds_full[['elu_150min','elu_180min','elu_210min','elu_240min','elu_270min','elu_300min','elu_330min','elu_360min','elu_390min']].melt()[['value']]
ds12 = ds_smaller.merge(ds1, how="left", on="NAME").sort_values(by=['T'])
ds13 = ds_small.merge(ds1, how="left", on="NAME").sort_values(by=['T'])
ds14 = ds_large.merge(ds1, how="left", on="NAME").sort_values(by=['T'])
ds1.to_csv('full_elu.csv', index=False)
ds12.to_csv('smaller_elu.csv', index=False)
ds13.to_csv('small_elu.csv', index=False)
ds14.to_csv('large_elu.csv', index=False)
```
